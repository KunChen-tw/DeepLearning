# Embedding model

```python
import torch

word1 = torch.LongTensor([0, 1, 100])

word2 = torch.LongTensor([2, 99, 3])

embedding = torch.nn.Embedding(101, 5)

print(embedding.weight)

print('word1:')

print(embedding(word1))

print('word2:')

print(embedding(word2))
```
